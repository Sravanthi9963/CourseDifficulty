# Data Preprocessing

```python
data.shape
```

```python
#Checking the Null values
data.isnull().sum()
```

```python
cols1=data._get_numeric_data().columns
cols2=df.columns
print(cols1)
```

```python
c1=['Termid', 'Regd No', 'CA_100', 'MTT_50', 'ETT_100', 'ETP_100',
       'Course_Att', 'CA_1', 'CA_2', 'CA_3', 'CA_4', 'Height', 'Weight']
c2=['Termid', 'Regd No', 'Course', 'Grade', 'CA_100', 'MTT_50', 'ETT_100',
       'ETP_100', 'Course_Att', 'MHRDName', 'CA_1', 'CA_2', 'CA_3', 'CA_4',
       'Height', 'Weight', 'ScholarType', 'Direction', 'Gender', 'Medium',
       'CourseType', 'ProgramType']
```

```python
#Categorical Features
c3=list(set(c2)-set(c1))
print(data[c3].isnull().sum())
```

```python
#Handling with the Null valued Columns: By replacing with the Mean of the column
from sklearn.impute import SimpleImputer
sm=SimpleImputer(strategy='mean',missing_values=np.nan)

for i in c1:
    data[i]=sm.fit_transform(data[[i]].values)
print(data.isnull().sum())
```

```python
data['Grade'].unique()
```

# Label Encoding

```python
#Label encoding of grade without using library
grade_map={'O':100,'A+':90,'A':80,'B+':70,'B':60,'C':50,'D':40,'E':30,'F':20,'I':10,'M':0,'R':0}
```

```python
data['Grade']=data['Grade'].map(grade_map)
data['Grade']
```

```python
courses=data['Course']
```

```python
#Label Encoding the Course feature using library
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder() 
data['Course']=le.fit_transform(data['Course'])
```

```python
dd=np.array(pd.DataFrame(data.groupby("Course",as_index=False)['Grade'].mean()))
di={}
for p in dd:
    di[p[0]]=p[1]
data['Strength']=data['Course'].map(di)
```

```python
#Since For predicting the difficulty of A Course i need features which have the correaltion with Course
##For training the data first convert the categorical to numerical where i used label encoding
import matplotlib.pyplot as plt
import seaborn as sns
li=['Course','CA_100','ETT_100','MTT_50','Strength']
X=data[li]  
sns.pairplot(X)
plt.show()
```

```python
#Splitting the data into train and test
X_train=X.sample(frac=0.75)
X_test = X.loc[~data.index.isin(X_train.index), :]
print(X_train.shape)
print(X_test.shape)
```

# K Mean Clustering

```python
from sklearn.cluster import KMeans
km=KMeans(n_clusters=3,max_iter=500,tol=0.001)
y_km=km.fit_predict(X_train)
```

```python
#Plotting the data which fall into their respective clusters
pp=np.array(X_train)
plt.scatter(pp[y_km==0,0],pp[y_km==0,1],c='red',marker='s',label='Easy')
plt.scatter(pp[y_km==1,0],pp[y_km==1,1],c='green',marker='o',label='Difficulty')
plt.scatter(pp[y_km==2,0],pp[y_km==2,1],c='blue',marker='*',label='Average')
plt.legend()
plt.grid()
plt.show()
```

```python
y_km
```

```python
km.cluster_centers_
```

```python
y_pred=km.predict(X_test)
```

```python
scores={}
from sklearn.metrics import silhouette_score
scores['KMean']=silhouette_score(X_train,y_km)
print(silhouette_score(X_train,y_km))
```

```python
#Checking the need for Paratameter optimizing (hyper parameter tuning) of n_clusters, 
import numpy as np
distortion=[]
for i in range(1,50,1):
    km=KMeans(n_clusters=i)
    y_km=km.fit_predict(X)
    distortion.append(km.inertia_)
 
```

```python
a=np.arange(1,50,1)
plt.plot(a,distortion)
plt.show()
```

# Agglomerative Clustering

```python
from sklearn.cluster import AgglomerativeClustering
clustering = AgglomerativeClustering(n_clusters=3).fit(X_train)
clustering
```

```python
y_ac=clustering.labels_
```

```python
pp=np.array(X_train)
plt.scatter(pp[y_ac==0,1],pp[y_ac==0,2],c='red',marker='s',label='cluster1')
plt.scatter(pp[y_ac==1,1],pp[y_ac==1,2],c='green',marker='o',label='cluster2')
plt.scatter(pp[y_ac==2,1],pp[y_ac==2,2],c='blue',marker='*',label='cluster3')
plt.legend()
plt.grid()
plt.show()
```

```python
from sklearn.metrics import silhouette_score
scores['Agglomerative']=silhouette_score(X_train,y_ac)
print(silhouette_score(X_train,y_ac))
```

# Clustering Using Birch

```python
from sklearn.cluster import Birch
model = Birch(threshold=0.01, n_clusters=3)
model.fit(X_train)
# assign a cluster to each example
y_bi = model.predict(X_train)
```

```python
pp=np.array(X_train)
plt.scatter(pp[y_bi==0,0],pp[y_bi==0,1],c='red',marker='s',label='Difficult')
plt.scatter(pp[y_bi==1,0],pp[y_bi==1,1],c='green',marker='o',label='Average')
plt.scatter(pp[y_bi==2,0],pp[y_bi==2,1],c='blue',marker='*',label='Easy')
plt.legend()
plt.grid()
plt.show()
```

```python
from sklearn.metrics import silhouette_score
scores['Birch']=silhouette_score(X_train,y_bi)
print(silhouette_score(X_train,y_bi))
```
